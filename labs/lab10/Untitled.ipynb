```python
def nonRepeating(s3):
    nonRepeat = []
    for n in range(0,len(s3)):
        for i in range(0,len(s3)):
            if s3[n] == s3[i]:
                 
            else:
                nonRepeat.append(s3[n])
                
    return nonRepeat
                
```
